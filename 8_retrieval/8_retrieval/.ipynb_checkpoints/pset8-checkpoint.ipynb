## Exercise 1

Write the function `make_constellation(snd, time_len, freq_len, plot)` which creates a point constellation from the spectrogram of the input audio data.

Inputs:
- `snd`: the audio data
- `time_len`: the zone length _in seconds_ for the 2D peak detection
- `freq_len`: the zone height _in Hertz_ for the 2D peak detection
- `plot`: optionally plot the spectrogram and points

Output:
- a $P \times 2$ matrix representing $P$ points where each row is the $(n,k)$ coordinates of that point.
 
Strategy:
- Create the magnitude spectrogram with parameters:
  - fft length: $N = 2048$
  - hop size: $H = 256$
  - You can use `fmp.stft` or `fmp.stft_mag`
- Only consider the frequency range $0-4000$Hz of the spectrogram. In other words, search only in a sub-range of rows of the spectrogram.
- To find peaks in the spectrogram, use the function `fmp.find_area_peaks`

```python
from scipy.ndimage.filters import maximum_filter
fft_len, hop_size, fs = 2048, 256, 22050.0
def f_to_k(f): return int((f*fft_len)/fs)
def t_to_samples(t): return t*fs/hop_size

def make_constellation(snd, time_len, freq_len, plot=False):
    lower_k, upper_k = f_to_k(0.0), f_to_k(4000.0)
    stft = fmp.stft(snd, win_len=fft_len,hop_size=hop_size)[lower_k:upper_k,:]
    peaks = fmp.find_area_peaks(img=np.abs(stft), \
                                zone_size=(f_to_k(freq_len), t_to_samples(time_len)))
    if plot:
        plt.figure()
        plt.imshow(np.abs(stft))
        plt.plot(peaks[:,0],peaks[:,1],'ro')
    return peaks
```

```python
snd = load_wav(filepath="./audio/beatles_magical_mystery_tour.wav", \
               t_start = 0, t_end=10)
time_len, freq_len, plot = 0.5, 400, True
peaks = make_constellation(snd, time_len, freq_len, plot)
```

## Exercise 5

Write the function `points_to_hash(points)` which converts a list of constellation points into hash lists with the hash function: $h=k$.

Inputs:
- `points`: a $P \times 2$ matrix of points where each row is the point $(n,k)$.

Output:
- A python dictionary where the keys are the various values of $h$ and the value of each key is a python list of the timestamps $n$.

If a hash key is not present in the points list, it should not have an entry in the dictionary.


```python
import itertools as it
import operator as op

def points_to_hash(points):
    return {k:reduce(op.add, map(lambda row:[row[0]],g)) \
            for k,g in it.groupby(points,key=lambda row:row[1])}
```

## Exercise 6

Write the function `hash_match(db_hash, query)` which finds the best timestamp that aligns the query points with the database points.

Inputs:
- `db_hash`: a dictionary representing the hash lists for the database
- `query`: a list of query points where each point is $(n,h)$. In this exercise, $h=k$ (the hash value of the point is the same as the frequency stamp).

Outputs:
- timestamp: the highest scoring timestamp for the query
- score: the score value (ie, number of point intersections when query is shifted by the timestamp).

Strategy:
- Use the matching algorithm as you did by hand in Exercise 4b. 
- Build up a list of $m$ values and find the highest occurring $m$.
- `np.unique` might be useful.

```python
def hash_match(db_hash, query):
    #db_hash = {h,[timestamp n]}, query = [(n,h)]
    h_to_m_list = {h: map(lambda l:(l-n), db_hash[h]) \
                    for n,h in query if h in db_hash}
    possible_ms = reduce(op.add, [m_list for h,m_list in h_to_m_list.iteritems()])
    m_counts = {m: sum([h_to_m_list[h].count(m) for h in h_to_m_list]) \
                    for m in possible_ms}
    timestamp,score = max(m_counts.iteritems(), key=op.itemgetter(1))
    return (timestamp,score)
```

## Exercise 7

Write the function `points_to_triplet_hash(points, return_list)` which create the more interesting hashing scheme. As described in the book and lecture, each item is calculated from a pair of points: an anchor point, and a fanout point. This function will be used to create either a database or a query from a list of points.

Inputs:
- `points`: a $P \times 2$ matrix of points where each row is the point $(n,k)$.
- `return_list`: if `False`, return a dictionary. if `True`, return a list.

Output:
- if `return_list == False`, return a python dictionary where the keys are the various values of $h$ and the value of each key is a python list of the timestamps $n$.
- if `return_list == True`, return a list of points $(n, h)$

Strategy:
- Iterate through all input points.
- For each point $(n_0, k_0)$, find all points that fall within the fanout zone. These are all points $(n_1, k_1)$ that satsify the conditions:
 - $n_1 > n_0$, $n_1 < n_0 + n_z$
 - $k_1 > k_0 - k_z$, $k_1 < k_0 + k_z$
 - Use the values $n_z = 200$ and $k_z = 40$
- The pairs created by $(n_0, k_0)$ and each $(n_1, k_1)$ are the items. Make the hash value ($h$) of each item be the python tuple $(k_0, k_1, n_1 - n_0)$. 
- Finally, the hash list values should be the timestamp of point $(n_0, k_0)$
- If returning a list, 

```python
nz,kz = 200,40

def pass_criteria(pt_pair):
    anchor_pt, target_pt = pt_pair
    n0,k0 = anchor_pt
    n1,k1 = target_pt
    return n0 < n1 < n0+nz and k0-kz < k1 < k0+kz

def hash_pair(pt_pair):
    anchor_pt, target_pt = pt_pair
    n0,k0 = anchor_pt
    n1,k1 = target_pt
    return (k0,k1,n1-n0)

def points_to_triplet_hash(points, return_list):
    pt_pairs = filter(pass_criteria, it.product(points,repeat=2))
    #points are [(n,k)]
    h_to_n_list = {h:reduce(op.add, map(lambda pt_pair:[pt_pair[0][0]], g)) \
                    for h,g in it.groupby(pt_pairs,key=lambda pt_pair:hash_pair(pt_pair))}
    #print h_to_n_list
    if return_list: 
        #pt_pairs  = [[(n,k),(n,k)]]
        hash_list = []
        for h in h_to_n_list:
            n_list = h_to_n_list[h]
            for n in n_list:
                hash_list.append((n,h))
        return hash_list
    return h_to_n_list
```

In this Exercise, we will explore audio matching. For simplicity, we will create a database of one song, though in general, the database is a concatenation of many songs.

The function `dtw_match_cost` below calculates warping paths that best match a query to a database by running dynamic time warping.

Inputs:
- `C`: the $N x M$ cost matrix that compares a query (length $N$) to a database (length $M$)

Outputs:
- `D`: the accumulated cost matrix
- `B`: the backtracking matrix. Each matrix cell is an index into the tuple `STEPS`


```python
STEPS = ((-1, -1), (-1, -2), (-2, -1))

def dtw_match_cost(C):
    N,M = C.shape
    
    # create accumulated cost matrix. Make D larger so that D[:,-1] and D[-1,:] are valid
    # and have inifinite cost. (this is due to STEPS having -2 in their step size)
    D = np.zeros((N+1, M+1))
    D[N,:] = 10000000
    D[:,M] = 10000000
    
    # backtracking index matrix - fill with zeros
    B = np.zeros(C.shape, dtype=np.int)

    # Initial values for D:
    D[0,0] = C[0,0]
    
    # first column
    for n in range(1,N):
        D[n,0] = C[n,0] + D[n-1,0]

    # first row - here we do not accumulate costs.
    D[0,0:M] = C[0,:]

    # compute accumulated cost, and keep track of steps
    for m in range(1,M):
        for n in range(1,N):
            # find minimal cost to reach (n,m). Accumulate into D. Remember step
            costs = (D[n-1,m-1], D[n-1,m-2], D[n-2,m-1])
            best_step  = np.argmin(costs)
            D[n,m] = costs[best_step] + C[n,m]
            B[n,m] = best_step

    # return accumulated cost and backtracking. 
    # Do not return that extra row/column of D.
    return D[:-1,:-1], B
```

## Exercise 9

Write the function `get_match_regions(d_dtw, B, num_matches)` that returns a list of matching paths calculated from the results of dynamic time warping.

Inputs:
- `d_dtw`: the matching function $\Delta_{\text{DTW}}(m)$
- `B`: the backtracking matrix
- `num_matches`: the desired number of matching paths to return

Output:
- a list of matching paths, where each path is a tuple of the start time, end time, and cost of that path

Strategy:
- Find the top N optimal endpoints of the matching function by using the function `fmp.find_top_n_troughs`. This is very similar to `find_top_n_peaks` you wrote in Lab, but it finds minima instead of maxima. For a half window length, use the length of the query itself.
- For each endpoint, find the startpoint by using the backtracking technique of dynamic time warping. Remember that each cell of $B$ is an index (0, 1 or 2) into the `STEPS` array.

```python
def traverse(end_position):
    y,x = end_position
    dy,dx = STEPS[B[y,x]]
    prev_position = (y+dy, x+dx)
    if y+dy==0 or x+dx==0: 
        return prev_position 
    return traverse(prev_position)

def get_match_regions(d_dtw, B, num_matches):
    N = B.shape[0]
    troughs = fmp.find_top_n_troughs(d_dtw,win_hlen=N,n=num_matches)
    plt.figure()
    plt.plot(d_dtw)
    plt.plot(troughs, d_dtw[troughs], 'ro')
    end_positions = map(lambda trough: (N-1,trough),troughs)
    start_positions = map(traverse, end_positions)
    print start_positions
    start_xs = map(lambda pt: pt[1], start_positions)
    plt.plot(start_xs, d_dtw[start_xs], 'go')
    costs = [d_dtw[t] for t in troughs]
    return zip(start_xs,troughs,costs)
    
    
```
